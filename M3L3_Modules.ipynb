```python
from time import *

start = time()
print(start)

sleep(5) # suspend program selama 5 detik

end = time()

total = round(end - start, 2) # dibulatkan jadi 2 angka di belakang koma
print(total)
```

Example

```python
from time import time

stopwatch = input('1 - start, 0 - stop:')

while stopwatch != '0':
  if stopwatch == '1':
      start = time()
  else:
      print('Action not found!')
  stopwatch = input('0 - stop:')

end = time()

total = end-start
print('Total time:', round(total, 2), 'c')
```
