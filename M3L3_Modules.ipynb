# Time module

image.png

```python
from time import *

time_now = time()

print(time_now)
```

```python
from time import *

start_time = time() # waktu start

print('haloo')

end_time = time() # waktu setelah print

total_time = end_time - start_time

print(total_time)
```

```python
from time import *

start_time = time() # waktu start

sleep(5) # sleep([jumlah detik])

end_time = time() # waktu setelah print

total_time = end_time - start_time

# dibulatkan 2 angka di belakang koma
# round([ yang mau dibulatkan], [angka di belakang koma])
print(round(total_time), 'detik')
```

```python
from time import *

start = time()
end = time()

total_time = end - start

print('total:', total_time, 'seconds')
```

```python
from time import *

start = time()

sleep(5) # suspend program for 5 seconds

end = time()

total_time = round(end - start) # round(decimal, jumlah angka di belakang koma)


print('total:', total_time, 'seconds')
```

```python
from time import *

start = time()
print(start)

sleep(5) # suspend program selama 5 detik

end = time()

total = round(end - start, 2) # dibulatkan jadi 2 angka di belakang koma
print(total)
```

Example

```python
from time import time

stopwatch = input('1 - start, 0 - stop:')

while stopwatch != '0':
  if stopwatch == '1':
      start = time()
  else:
      print('Action not found!')
  stopwatch = input('0 - stop:')

end = time()

total = end-start
print('Total time:', round(total, 2), 'c')
```
