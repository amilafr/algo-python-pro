## Without Parameter (Tanpa parameter)

```python
def print_data():
  print('My name is John')
  print('I am from UK')
  print('I am 20 yo')

print_data()
```

```python
for i in range(3):
  print_data()
```

```python
def print_label():
       print('THE SUCCESS CENTER')
       print('Name: ____')
       print('Course: ____')
       print('Group: ______')

print('Label printing machine')
amount = int(input('Number of students:'))
for i in range (amount):
  print_label()
print('Done! Take your labels.')
```

## With Parameter (dengan parameter)

Argument --> data yang dimasukkan sebagai parameter

```python
def print_data(name, place, age):
  print('My name is', name)
  print('I am from', place)
  print('I am', age, 'yo')

print_data('alex', 'us', 17)
```

```python
def print_label(name):
   print('THE SUCCESS CENTER')
   print('Name:', name)
   print('Course: English')

print('Test worksheets printing')
amount = int(input('Number of students:'))
for i in range(amount):
   name = input('Student name:')
   print_label(name)
print('Done! Take the sheets.')
```

# Returning a Value (mengembalikan nilai)

image.png

image.png

```python
def get_result(score):
  if score >= 50 and score < 100:
      result = 'grade III certificate'
  elif score >= 100 and score < 200:
      result = 'grade II certificate'
  elif score >= 200:
      result = 'grade I certificate'
  else:
      result = 'appreciation letter'
  return result

name = input('Student name:')
score = int(input('Points earned:'))
result = get_result(score)
print(name, '-', result)
if result == 'grade I certificate':
   print('You receive a gift — a book shop voucher ')
```

## Mengembalikan Nilai

```python
def penjumlahan(num1, num2):
  hasil = num1 + num2
  return hasil # ada return --> mengembalikan nilai ke 'hasil'

# cara memanggil
result = penjumlahan(2, 3) # disimpan dalam variabel
print(result)
```

```python
print(penjumlahan(2, 3)) # langsung
```

```python
# kalau ada if-else, return menyesuaikan cabang
def check_score(score):
    if score >= 80:
        return True # ada return
    else:
        return False

score_result = check_score(70)
print(score_result)
```

```python
score = int(input('Masukkan skor:'))

score_result = check_score(score)
print(score_result)
```
