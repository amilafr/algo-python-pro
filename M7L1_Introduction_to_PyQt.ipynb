# Winner Identifier

image.png

```python
# import --> menghubungkan library PyQt5
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import QApplication, QWidget, QPushButton, QLabel, QVBoxLayout
from random import randint

# bikin objek aplikasi dan window nya
app = QApplication([])
main_win = QWidget()

# screen = app.primaryScreen()
# size = screen.size()

main_win.setWindowTitle('Penentu pemenang')
main_win.resize(400, 200) # lebar, tinggi
# main_win.move(900, 500) # x, y
# main_win.setGeometry()

# membuat widget
but = QPushButton('Generate')
text = QLabel('Klik untuk mengetahui pemenang')
winner = QLabel('?')

line = QVBoxLayout() # layout vertical

# menempatkan widget pada layout
line.addWidget(text, alignment=Qt.AlignCenter)
line.addWidget(winner, alignment=Qt.AlignCenter)
line.addWidget(but, alignment=Qt.AlignCenter)

main_win.setLayout(line)

# Jika hanya ada 100 peserta, maka dihasilkan angka acak dari 1 hingga 100.
# Label 'teks' berubah dari "Klik untuk mengetahui pemenang" menjadi "Pemenang".
# Label 'pemenang' berubah dari "?" ke nomor pemenang.

# event handling
def show_winner():
    number = randint(1, 100)
    winner.setText(str(number))
    text.setText('Pemenang')

but.clicked.connect(show_winner)

# meluncurkan aplikasi
main_win.show() # menampilkan window
app.exec_() # aplikasinya muncul terus sampai klik exit
```

# Quiz

```python
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import QApplication, QWidget, QRadioButton, QPushButton, QLabel, QVBoxLayout, QHBoxLayout, QMessageBox

# bikin objek aplikasi dan window nya
app = QApplication([])
main_win = QWidget()

main_win.setWindowTitle('Quiz')
main_win.resize(400, 200) # lebar, tinggi

question = QLabel('Kapan Algorithmics didirikan?')
ans1 = QRadioButton('2013')
ans2 = QRadioButton('2010')
ans3 = QRadioButton('2016')
ans4 = QRadioButton('2019')

# layout
vlayout = QVBoxLayout()

layoutH1 = QHBoxLayout()
layoutH2 = QHBoxLayout()
layoutH3 = QHBoxLayout()

# horizontal
layoutH1.addWidget(question, alignment=Qt.AlignCenter)

layoutH2.addWidget(ans1, alignment=Qt.AlignCenter)
layoutH2.addWidget(ans2, alignment=Qt.AlignCenter)

layoutH3.addWidget(ans3, alignment=Qt.AlignCenter)
layoutH3.addWidget(ans4, alignment=Qt.AlignCenter)

# vertical
vlayout.addLayout(layoutH1)
vlayout.addLayout(layoutH2)
vlayout.addLayout(layoutH3)

main_win.setLayout(vlayout)

# benar dan salah
def show_win():
    winner = QMessageBox()
    winner.setText('Benar')
    winner.exec_()

def show_lose():
    lose = QMessageBox()
    lose.setText('Salah')
    lose.exec_()

ans3.clicked.connect(show_win)
ans1.clicked.connect(show_lose)
ans2.clicked.connect(show_lose)
ans4.clicked.connect(show_lose)

# meluncurkan aplikasi
main_win.show() # menampilkan window
app.exec_() # aplikasinya muncul terus sampai klik exit


# vlayout.addWidget(question, alignment=Qt.AlignCenter)
# vlayout.addWidget(ans1, alignment=Qt.AlignCenter)
# vlayout.addWidget(ans2, alignment=Qt.AlignCenter)
# vlayout.addWidget(ans3, alignment=Qt.AlignCenter)
# vlayout.addWidget(ans4, alignment=Qt.AlignCenter)
```
