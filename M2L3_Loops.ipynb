# Loop : While

**Repeat until** --> gatau berapa kali perulangannya

image.png

```python
# while --> berulang terus selama kondisi True

userInput = input('Enter your answer: ')

while userInput != 'off':
  print('hello')
  userInput = input('Enter your answer: ')

print('The end.')
```

```python
price = int(input('Enter the price (0 - finish): '))
total = 0

while price != 0:
  total += price
  price = int(input('Enter the price (0 - finish): '))

print('Total price:', total)
```

```python
price = int(input('Enter the next price (0 to finish):'))
total_price = 0

while price != 0:
   total_price += price
   price = int(input('Enter the next price (0 to finish):'))

print('Total price without discount:', total_price)

total_price = total_price * 0.9
print('Total price with discount:', total_price)
```

# Loop with Counter

**Counter** --> this is a variable storing the number of steps of a certain loop.


image.png

```python
promo = input('Promo code: ')
counter = 0

while promo != 'life':
  print('Error')
  promo = input('Promo code: ')
  counter += 1


print('Promo accepted.')
print('Jumlah loop:', counter)
```

```python
counter = 0

while counter < 5:
  print('hello')
  counter += 1

print('the end.')
```

```python
password = input('Password: ')
counter = 1

while password != 'life':
  print('attempt', counter)
  print('error')
  password = input('Password: ')
  counter += 1
```

```python
password = input('Input password: ')
attempts = 1

while password != 'hola':
  print('Error')
  password = input('Input password: ')
  attempts += 1

print('Total attempts:', attempts)
```

```python
# rata-rata
total = 0
counter = 0

angka = int(input('Masukkan angka (0 - stop): '))

while angka != 0:
  total += angka
  counter += 1
  print('jumlah angka:', counter)
  angka = int(input('Masukkan angka (0 - stop): '))

mean = total / counter
print('hasilnya:', mean)
```
