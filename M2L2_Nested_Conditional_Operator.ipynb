```python
wish = input('What do you want to eat?')

if wish == 'cake':
  flavor = input('Which one do you want?')
  if flavor == 'choclocate':
    print('Chocolate cake')
  else:
    print('Mango cake')

else: # otherwise
  wish2 = input('Which one do you want?')
  if wish2 == 'rice':
    print('Yey rice')
```

```python
# nilai 90 - 100 --> A
# nilai 80 - 89 --> B
# lainnya itu C
nilai = 80

if nilai >= 90 and nilai <= 100: # 90 - 100
  print('Nilai A')
elif nilai >= 80 and nilai <= 89: # 80-89
  print('B')
  if nilai == 80:
    print()
  else:
    print()
else:
  print('C')
```
