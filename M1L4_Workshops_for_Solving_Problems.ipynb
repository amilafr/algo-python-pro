```python
# 6 - The program after optimization (other options are possible, in which the program takes the same number of lines or less)

month_salary = int(input('Enter the monthly salary:'))
vacation = int(input('Enter the number of vacation days:'))

daily_salary = month_salary / 29.3 #average number of days in a month
vacation_pay = daily_salary * vacation

print('Approximate vacation pay:', vacation_pay)
```

```python
# 7 - The program for awarding the award:

salary = int(input('Monthly salary:'))
extra_time = int(input('Number of hours worked on weekends:'))

bonus = salary * 0.01 * extra_time

print('Premium size:', bonus)
```
