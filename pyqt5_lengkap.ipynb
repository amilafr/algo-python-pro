# my_app.py --> halaman pertama (file utama)

```python
# write here a code for the main app and the first screen
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import (
       QApplication, QWidget,
       QHBoxLayout, QVBoxLayout,
       QGroupBox, QRadioButton,
       QPushButton, QLabel, QListWidget, QLineEdit)

from instr import *
from second_win import TestWin

class MainWin(QWidget):
    # constructor
    def __init__(self):
        super().__init__()
        self.set_appear() # tampilan window
        self.initUI() # widget2
        self.connects() # koneksi antar window
        self.show() # run/start

    def set_appear(self):
        self.setWindowTitle(txt_title) # judul app
        self.resize(win_width, win_height)
        self.move(win_x, win_y)

    # elemen grafis --> widget2
    def initUI(self):
        self.hello_text = QLabel(text=txt_hello)
        self.instruction = QLabel(text=txt_instruction)
        self.button = QPushButton(text=txt_next)

        # layout
        self.v_layout = QVBoxLayout()
        # tambahin widget ke layout
        self.v_layout.addWidget(self.hello_text, alignment=Qt.AlignLeft)
        self.v_layout.addWidget(self.instruction, alignment=Qt.AlignLeft)
        self.v_layout.addWidget(self.button, alignment=Qt.AlignCenter)

        self.setLayout(self.v_layout)

    def connects(self):
        self.button.clicked.connect(self.next_click)

    def next_click(self):
        self.hide() # hilang
        self.tw = TestWin()

# run app
app = QApplication([]) # bikin app
mw = MainWin() # window
app.exec_() # app run terus sampai klik close
```

```python
# write a code for the second app
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import (
       QApplication, QWidget,
       QHBoxLayout, QVBoxLayout,
       QGroupBox, QRadioButton,
       QPushButton, QLabel, QListWidget, QLineEdit)

from instr import *
from final_win import FinalWin

class Experiment():
    def __init__(self, age, test1, test2, test3):
        self.age = age
        self.test1 = test1
        self.test2 = test2
        self.test3 = test3

class TestWin(QWidget):
    def __init__(self):
        super().__init__()
        self.set_appear() # tampilan window
        self.initUI() # widget2
        self.connects() # koneksi antar window
        self.show() # run/start

    def set_appear(self):
        self.setWindowTitle(txt_title) # judul app
        self.resize(win_width, win_height)
        self.move(win_x, win_y)

    def initUI(self):
        self.name_txt = QLabel(text=txt_name)
        self.in_name = QLineEdit(txt_hintname)

        self.year_txt = QLabel(text=txt_age)
        self.in_age = QLineEdit('0')

        self.test1_txt = QLabel(text=txt_test1)
        self.btn_test1 = QPushButton(text=txt_starttest1)

        self.in_pulse1 = QLineEdit('0')

        self.test2_txt = QLabel(text=txt_test2)
        self.btn_test2 = QPushButton(text=txt_starttest2)

        self.test3_txt = QLabel(text=txt_test3)
        self.btn_test3 = QPushButton(text=txt_starttest3)

        self.in_pulse2 = QLineEdit('0')
        self.in_pulse3 = QLineEdit('0')

        self.btn = QPushButton(text='Send the results')

        # layout
        self.v_layout = QVBoxLayout()
        self.v_layout.addWidget(self.name_txt, alignment=Qt.AlignLeft)
        self.v_layout.addWidget(self.in_name, alignment=Qt.AlignLeft)
        self.v_layout.addWidget(self.year_txt, alignment=Qt.AlignLeft)
        self.v_layout.addWidget(self.in_age, alignment=Qt.AlignLeft)
        self.v_layout.addWidget(self.test1_txt, alignment=Qt.AlignLeft)
        self.v_layout.addWidget(self.btn_test1, alignment=Qt.AlignLeft)
        self.v_layout.addWidget(self.in_pulse1, alignment=Qt.AlignLeft)
        self.v_layout.addWidget(self.test2_txt, alignment=Qt.AlignLeft)
        self.v_layout.addWidget(self.btn_test2, alignment=Qt.AlignLeft)
        self.v_layout.addWidget(self.test3_txt, alignment=Qt.AlignLeft)
        self.v_layout.addWidget(self.btn_test3, alignment=Qt.AlignLeft)
        self.v_layout.addWidget(self.in_pulse2, alignment=Qt.AlignLeft)
        self.v_layout.addWidget(self.in_pulse3, alignment=Qt.AlignLeft)
        self.v_layout.addWidget(self.btn, alignment=Qt.AlignCenter)

        self.setLayout(self.v_layout)

    def connects(self):
        self.btn.clicked.connect(self.next_click)

    def next_click(self):
        self.hide() # hilang

        self.exp = Experiment(self.in_age.text(), self.in_pulse1.text(), self.in_pulse2.text(), self.in_pulse3.text())

        self.fw = FinalWin(self.exp)

```

```python
# write here a code for the third app
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import (
       QApplication, QWidget,
       QHBoxLayout, QVBoxLayout, QGridLayout,
       QGroupBox, QRadioButton,
       QPushButton, QLabel, QListWidget, QLineEdit)

from instr import *

class FinalWin(QWidget):
    # constructor
    def __init__(self, exp):
        super().__init__()

        self.exp = exp

        self.set_appear() # tampilan window
        self.initUI() # widget2
        # self.connects() # koneksi antar window
        self.show() # run/start

    def set_appear(self):
        self.setWindowTitle(txt_finalwin) # judul app
        self.resize(win_width, win_height)
        self.move(win_x, win_y)

    # elemen grafis --> widget2
    def initUI(self):
        self.text2 = QLabel(text=txt_workheart + self.results())
        self.text1 = QLabel(text=txt_index + str(self.index))

        # layout
        self.v_layout = QVBoxLayout()
        # tambahin widget ke layout
        self.v_layout.addWidget(self.text1, alignment=Qt.AlignCenter)
        self.v_layout.addWidget(self.text2, alignment=Qt.AlignCenter)

        self.setLayout(self.v_layout)

    def results(self):
        self.index = (4 * (int(self.exp.test1) + int(self.exp.test2) + int(self.exp.test3)) -200) /10

        if int(self.exp.age) >= 15:
            if self.index >= 15:
                return txt_res1
            elif self.index >= 11 and self.index < 15:
                return txt_res2
            elif self.index >= 6 and self.index < 11:
                return txt_res3
            elif self.index >= 0.5 and self.index < 6:
                return txt_res4
            else:
                return txt_res5

    # def connects(self):
    #     self.button.clicked.connect(self.next_click)

    # def next_click(self):
    #     self.hide() # hilang
    #     self.tw = TestWin()

```
