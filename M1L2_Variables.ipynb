# Variable

Variabel adalah elemen data yang memiliki nama. Variabel digunakan untuk bekerja dengan data yang dapat berubah.

image.png

```python
angka1 = 4
angka2 = 7

hasil = angka1 * angka2

print('Hasil dari', angka1, 'dikali', angka2, 'adalah', hasil)
```

image.png

```python
# Tipe data variabel
variabel_integer = 5 #bilangan bulat
variabel_float = 5.5 #bilangan desimal
variabel_string = 'haloo'

print(variabel_integer)
print(variabel_float)
print(variabel_string)
```

```python
angka = 6
print(angka)
```

```python
angka = 6
angka = 7.5 # mengubah nilai variabel angka jadi 7.5

print(angka)
```

# Fungsi int() dan str()

image.png

```python
jumlah = input('Masukkan jumlah:')
jumlah = int(jumlah) # mengubah jadi integer

total = jumlah + 1000

print('Totalnya adalah', total)
```

```python
angka = 100
angka = str(angka) # mengubah jadi string

hasil = 'hasilnya ' + angka

print(hasil)
```
