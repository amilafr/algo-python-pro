## Fungsi int() dan str()

image.png

```python
# fungsi input --> string "5"
umur = input("Masukkan umur : ")

# umur diubah jadi integer
umur = int(umur)

print(umur*2) # "20" --> 40
```

```python
jumlah = input('Masukkan jumlah:')
jumlah = int(jumlah) # mengubah jadi integer

total = jumlah + 1000

print('Totalnya adalah', total)
```

```python
angka = 100 # integer
angka = str(angka) # mengubah jadi string

hasil = "angkanya adalah " + angka

print(hasil)
```
