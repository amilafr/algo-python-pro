## Contoh --> perulangan untuk String

```python
username = 'anny123'

# a n n y 1 2 3

for elemen in username:
  print(elemen)
```

```python
kata = 'malam'

for huruf in kata:
  print(huruf, '...')
```

```python
kata = 'selamat' # s, e, l, a, m, a, t

for e in kata:
  print(e)
```

```python
kata = 'halo'

for huruf in kata:
  print(huruf)
```

```python
kata = 'halo'
alphabet = 'abcdefghijklmnopqrstuvwxyz'

for huruf in kata:
  cari = alphabet.find(huruf) + 1
  print(huruf, cari)
```

```python
password = 'hello'
alphabet = 'abcdefghijklmnopqrstuvwxyz'

for symbol in password:
  symbol_number = alphabet.find(symbol) + 1
  print(symbol_number)
```

```python
login = input('Enter your login:')
wrong = '=?*^$@_'

for symbol in login:
   if symbol in wrong:
       print('Forbidden symbol:', symbol)
```

```python
kata = input('masukkan kata: ')
terlarang = '@#$'

for elemen in kata:
  if elemen in terlarang: # elemen di kataada di terlarang
    print(elemen)
```
