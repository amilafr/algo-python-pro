Contoh --> perulangan untuk String

```python
kata = 'halo'

for huruf in kata:
  print(huruf)
```

```python
password = 'hello'
alphabet = 'abcdefghijklmnopqrstuvwxyz'

for symbol in password:
  symbol_number = alphabet.find(symbol) + 1
  print(symbol_number)
```

```python
login = input('Enter your login:')
wrong = '=?*^$№@_'

for symbol in login:
   if symbol in wrong:
       print('Forbidden symbol:', symbol)
```

```python
for i in range(4): # mengulang sebanyak 4x --> 0-3
  print(i, 'halo')
```

```python
for i in range(5, 10): # mengulang dari 5 sampai <10 --> 5-9
  print(i, 'hai')
```

```python
amount = int(input('Number of participants:'))

for i in range(amount):
   name = input('Enter name:')
   print('Welcome,', name)

print('Group chat is created!')
```
