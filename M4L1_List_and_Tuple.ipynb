## Example 2

```python
# mencari rata2 dan max number

results = [225, 220, 199, 263, 259, 225, 226]
average_result = 0
max_result = 0

for result in results:
  average_result += result
  if max_result < result:
      max_result = result

average_result = average_result / len(results)

print('Hasil:', results)
print('Hasil rata-rata:', average_result)
print('Hasil maksimum:', max_result)
```

```python
numbers = ['5', '6', '7']
print(numbers)

numbers = list(map(int, numbers)) # untuk menerapkan fungsi ke semua elemen di list
print(numbers)
```

```python
numbers = input('Masukkan angka: ') # String
numbers = numbers.split(' ') # list --> string

print(numbers)

numbers = list(map(int, numbers))

print(numbers)

total = 0
for num in numbers:
  total += num

print(total)
```
