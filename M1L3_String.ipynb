## **Lowercase** --> method .lower()

string_var.lower()

image.png

```python
sentence = "I love coding"

search = sentence.find('i love')

print(search)
```

```python
# method .lower() --> semua capital jadi lowercase
sentence = "I love coding"
sentence = sentence. lower() # lowercase all characters --> string_variable.lower()

search = sentence.find('i love')

print(sentence)
print(search)
```

```python
kalimat = "Aku belajar python"
kalimat = kalimat.lower() # bikin semua hurufnya jadi huruf kecil

cari = kalimat.find('aku')

print(cari)
```

```python
#bikin semua hurufnya jadi kecil (lowercase) --> method .lower()
kalimat = kalimat.lower()
kata_aku = kalimat.find('aku')

print(kata_aku)
```

```python
kalimat = 'Aku sedang Belajar'
print(kalimat)

kalimat = kalimat.lower()
print(kalimat)

cari = kalimat.find('belajar')

print(cari)
```

## **Strange program**

```python
rate1 = input('masukkan rating1:')
rate2 = input('masukkan rating2:')

total = rate1 + rate2
print(total)

# string + string --> merge (gabung) 2 string
```

## String --> Integer

```python
# mengubah string -> int

rate1 = int(rate1)# mengubah tipe data jadi int
rate2 = int(rate2)

total = rate1 + rate2
print(total)
```

## Operator +

```python
total = 4 + 5

print(total)
```

```python
total = 'hai' + 'peter'

print(total)
```

```python
rate1 = 4
rate2 = 5

total = rate1 + rate2
print(total)

# integer + integer --> penjumlahan
```

```python
# string + integer --> error
a = 'halo'
b = 5
b = str(5) # mengubah tipe data jadi string

hasil = a + b
print('pake +:', hasil)

print(a, b)
```

## Operator *

```python
hasil = 5 * 3

print(hasil)
```

```python
hasil = 'aku' * 'kamu'

print(hasil)
```

```python
hasil = 'aku' * 3

print(hasil)
```

```python
# string * int --> ngulang string
hasil = 3 * 'hai' # hai jadi 3x
print(hasil)

# int * int --> perkalian
hasil = 3 * 3
print(hasil)
```

```python
#string * string --> error
hasil = 'a' * 'b'
print(hasil)
```
