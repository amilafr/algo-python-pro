## **Slicing String**

memotong string dari index a sampai < b

```python
# nyari pake rentang --> slicing
kalimat = "Aku belajar python"

# Aku belajar
slicing1 = kalimat[0:11] # slicing dari index 0 sampai index < 10

print(slicing1)
```

## **Find in String** --> method .find()

mencari kata di string

string_var.find(word)

```python
kalimat = "Aku belajar python"

# ngecek kata python itu ada gak di stringnya
cari = kalimat.find('belajar')
print(cari) # hasilnya kalo ada --> index huruf pertama yg dicari
```

```python
cari = kalimat.find('aku')
print(cari)
```

```python
#nyari kata di string

kalimat = "Aku belajar python"
# nyari kata python
cari = kalimat.find('belajar')

print(cari) # 4 --> index huruf b
```

```python
# kalo tidak ada
cari = kalimat.find('minggu')
print(cari)
```

**Lowercase** --> method .lower()

string_var.lower()

```python
kalimat = "Aku belajar python"
kalimat = kalimat.lower() # bikin semua hurufnya jadi huruf kecil

cari = kalimat.find('aku')

print(cari)
```

```python
#bikin semua hurufnya jadi kecil (lowercase) --> method .lower()
kalimat = kalimat.lower()
kata_aku = kalimat.find('aku')

print(kata_aku)
```

```python
rate1 = input('masukkan rating1:')
rate2 = input('masukkan rating2:')

total = rate1 + rate2
print(total)

# string + string --> merge (gabung) 2 string
```
