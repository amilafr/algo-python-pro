#String


```python
kalimat = 'Aku senang belajar python'
#index dimulai dari 0 --> 'A'
```

```python
#cari tahu panjang dari string (jumlah karakter) --> fungsi len()
panjang_kalimat = len(kalimat)
print(panjang_kalimat)
```

```python
#get karakter di index tertentu
cari = kalimat[1] # nama_variabel[index]
print(cari)
```

```python
#nyari pake rentang --> slicing
slicing = kalimat[0:6] #index 0 sampai <3
print(slicing)
```

```python
#nyari kata di string
search = kalimat.find('python')
print('python:', search) #output --> index karakter pertama ('p')

gaada = kalimat.find('pagi')
print('kalo gaada:', gaada)
```

```python
#bikin semua hurufnya jadi kecil (lowercase) --> method .lower()
kalimat = kalimat.lower()
kata_aku = kalimat.find('aku')
print(kata_aku)
```

```python
rate1 = input('masukkan rating1:')
rate2 = input('masukkan rating2:')

total = rate1 + rate2
print(total)

# string + string --> merge (gabung) 2 string
```

```python
rate1 = 4
rate2 = 5

total = rate1 + rate2
print(total)
# integer + integer --> penjumlahan
```

```python
# string + integer
a = 'halo'
b = 5
b = str(5) # mengubah tipe data jadi string

hasil = a + b
print('pake +:', hasil)

print(a, b)
```

```python
# string * int --> ngulang string
hasil = 3 * 'hai' #hai jadi 3x
print(hasil)

# int * int --> perkalian
hasil = 3 * 3
print(hasil)
```

```python
#string * string --> error
hasil = 'a' * 'b'
print(hasil)
```
