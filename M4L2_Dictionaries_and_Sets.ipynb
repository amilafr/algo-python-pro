```python
buku = {
    'horror' : 'The Conjuring',
    'romance' : 'Dilan'
    }

print(buku['horror'])
```

image.png

```python
# Nyari data --> pake kata kunci

search = buku['romance'] # dict_variable[key]
print(search)
```

```python
# Panjang dictionaries

length = len(buku) # sama kyk list --> len()
print(length)
```

```python
# Get keys --> ambil semua keys di dictionary

print(buku.keys()) # dict_variable.keys()
```

```python
# Get values --> ambil semua values di dictionary

print(buku.values()) # dict_variable.values()
```

```python
# Nambahin data -- key + value in dictionary

# nambah data --> key = 'non fiction', value = 'Michael Jackson'
buku['nonfiction'] = 'Michael Jackson'

print(buku)
```

```python
# Delete key and its value

del buku['horror']
```

```python
print(buku)
```

```python
# Nyari key --> True or False

cari_horror = 'horror' in buku
print(cari_horror)

cari_romance = 'romance' in buku
print(cari_romance)
```

#Sets
**An unordered set of unique elements.**

setiap data yang ada tidak boleh sama

image.png

```python
numbers_list = [1, 2, 2, 1, 1, 1]

print(numbers_list)
```

```python
numbers = set()
numbers = {1, 2, 2, 1, 1, 1}

print(numbers)
```

```python
# Length of a Set

length = len(numbers) # sama

print(length)
```

```python
numbers_set = set(numbers_list)

print(numbers_list)
print(numbers_set)
```

```python
# Bikin Set dari Data Structure lain

# List --> Set
my_list = [1, 2, 3, 3, 4, 2, 1]
my_set = set(my_list) # set(nama_list)

print(my_list)
print('panjang list:', len(my_list))

print(my_set)
print('panjang set:', len(my_set))
```

image.png

image.png

```python
# Add and Remove Element in Set

set_number = {1, 2, 3}

set_number.add(5) # nambah data
print(set_number)
```

```python
# remove
set_number.remove(3) # nama_set.remove(value)

print(set_number)
```

```python
# clear all elements
set_number.clear() # kehapus semua --> set kosong

print(set_number)
```

```python
# Search element in Set --> True or False

my_set = {1, 2, 3}

# cek elemen
check_set = 2 in my_set # sama kyk dictionary
print(check_set)
```

image.png

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
```

```python
# Union --> gabungan set A dan B
union_ab = A.union(B)

print(union_ab)
```

```python
# Intersection --> Irisan (Nilai yang sama)
inter_ab = A.intersection(B)

print(inter_ab)
```

```python
# Difference --> yang gak sama
diff_ab = A.difference(B) # ada di A tapi gak ada di B

print(diff_ab)

print(B.difference(A)) # ada di B gak ada di A
```

```python
authors = {
   'Kipling' : 'Novelis Inggris, penulis cerita pendek, penyair, dan jurnalis. Salah satu tokoh sastra paling otoritatif, dia adalah salah satu penulis paling populer di Britania Raya pada akhir abad ke-19 dan awal abad ke-20',
   'Tolkien': 'Penulis Inggris, penyair, filolog, dan akademisi, paling dikenal sebagai penulis karya fantasi The Hobbit dan The Lord of the Rings.',
   'Camus' : 'Filusuf, penulis, dramawan, dan jurnalis Perancis. Dianugerahi Hadiah Nobel untuk Sastra pada tahun 1957'}


surname = input('Nama belakang penulis:')
if surname in authors:
   print('Biografi:', authors[surname])
else:
    answer = input('Penulis tidak ditemukan. Ingin menambahkan?')
    answer = answer.lower()
    if answer == 'ya':
        biography = input('Biografi penulis:')
        authors[surname] = biography
        print(authors)
    else:
        print('Tanggapan diterima')
```

```python
student_card = {'angka': '324', 'nama belakang': 'Smith'}
print('Selamat datang!', student_card)

answer = int(input('Akun pribadi: 1 - ambil buku, 2 - kembalikan buku, 3 - pulang'))

while answer != 3:
  if answer == 1:
      title = input('Masukkan judul:')
      student_card['utang'] = title
      print('Kartu pembaca:', student_card)
  if answer == 2:
      if 'utang' in student_card:
          del student_card['utang']
      print('Kartu pembaca:', student_card)
  answer = int(input('Tindakan Anda: 1 - ambil buku, 2 - kembalikan buku, 3 - pulang'))

print('Sampai jumpa lagi:', student_card)
```

## 2.5

```python
def application():
    contestants = dict()
    name = input('Masukkan nama peserta (0 - lengkap)')
    while name != '0':
        poem = input('Masukkan judul karya')
        contestants[name] = poem
        name = input('Masukkan nama peserta (0 - lengkap)')
    return contestants

def get_artists(poem):
    artists = []
    for artist in contestants:
        if contestants[artist] == poem:
            artists.append(artist)
        artists.sort()
    return '(' + ', '.join(artists) + ')'

contestants = application()
concert_program = set(contestants.values())
print('PROGRAM KONSER')
for item in concert_program:
    print(item, get_artists(item))
```
