#Dictionaries

**An unordered set of key-value pairs.**

cth : nyari buku berdasarkan genre

```python
# Deklarasi Dictionary

books = dict() # dictionary kosong
books = {'romance' : 'You and Me',
         'horror' : 'Ghost Hunter',
         'comedy' : 'Laugh Out Loud'}

print(books)
```

image.png

```python
# Get dictionary value by 'key'

search = books['romance'] # dict_variable[key]
print(search)
```

```python
# Length of Dictionaries

length = len(books) # sama kyk list --> len()
print(length)
```

```python
# Get keys --> ambil semua keys di dictionary

print(books.keys()) # dict_variable.keys()
```

```python
# Get values --> ambil semua values di dictionary

print(books.values()) # dict_variable.values()
```

```python
# Add key + value in dictionary

# nambah data --> key = 'non fiction', value = 'Michael Jackson'
books['non fiction'] = 'Michael Jackson'

print(books)
print('length:', len(books))
```

```python
# Search for the occurence of a key --> True or False

genre1 = 'non fiction' in books
genre2 = 'adventure' in books

print('non fiction:', genre1)
print('adventure:', genre2)
```

```python
# Delete key and its value

del books['horror']

print(books)
print('length:', len(books))
```

#Sets
**An unordered set of unique elements.**

setiap data yang ada tidak boleh sama

```python
# Deklarasi Set

numbers = set() # set kosong
numbers = {1, 2, 3}

print(numbers)
```

```python
# Length of a Set

length = len(numbers) # sama

print(length)
```

```python
# Bikin Set dari Data Structure lain

# List --> Set
my_list = [1, 2, 3, 3, 4, 2, 1]
my_set = set(my_list) # set(nama_list)

print(my_list)
print('panjang list:', len(my_list))

print(my_set)
print('panjang set:', len(my_set))
```

image.png

```python
# Add and Remove Element in Set

# add
my_set.add(8) # nama_set.add(value)
# my_set.update(6)

print(my_set)
```

```python
# remove
my_set.remove(3) # nama_set.remove(value)

print(my_set)
```

```python
# clear all elements
my_set.clear() # kehapus semua --> set kosong

print(my_set)
```

```python
# Search element in Set --> True or False

my_set = {1, 2, 3}

# cek elemen
check_set = 2 in my_set # sama kyk dictionary
print(check_set)
```

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
```

```python
# Union --> gabungan set A dan B
union_ab = A.union(B)

print(union_ab)
```

```python
# Intersection --> Irisan (Nilai yang sama)
inter_ab = A.intersection(B)

print(inter_ab)
```

```python
# Difference --> yang gak sama
diff_ab = A.difference(B)

print(diff_ab)

print(B.difference(A))
```
