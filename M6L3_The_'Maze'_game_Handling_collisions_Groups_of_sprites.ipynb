```python
from pygame import *

window = display.set_mode((700, 500))
display.set_caption('Maze Game')

bg_color = (194, 236, 254)

'''class parent'''
class GameSprite(sprite.Sprite):
    def __init__(self, img, width, height, x, y):
        sprite.Sprite.__init__(self)
        self.image = transform.scale(image.load(img), (width, height))
        self.rect = self.image.get_rect()
        # position
        self.rect.x = x # x coordinate
        self.rect.y = y # y coordinate

    # placing object on the screen
    def reset(self):
        window.blit(self.image, (self.rect.x, self.rect.y))

# class player
class Player(GameSprite):
    def __init__(self, img, width, height, x, y, x_speed, y_speed):
        GameSprite.__init__(self, img, width, height, x, y)
        self.x_speed = x_speed
        self.y_speed = y_speed

    #  # movement
    def update(self):
        self.rect.x += self.x_speed
        self.rect.y += self.y_speed


class Rectangle(sprite.Sprite):
    # constructor
    def __init__(self, width, height, x, y, color):
        super().__init__()
        self.rect = Rect(x, y, width, height)
        self.fill_color = color

    # placing object on the screen
    def draw(self):
        draw.rect(window, self.fill_color, self.rect)

class Circle(sprite.Sprite):
    def __init__(self, center, radius, color='red'):
        super().__init__()
        self.center = center
        self.radius = radius
        self.color = color
    # placing object on the screen
    def draw(self):
        draw.circle(window, self.color, self.center, self.radius)

# objects
wall1 = GameSprite('platform2.png', 300, 50, 100, 100)
player = Player('hero.png', 80, 80, 100, 200, 0, 0)
rectangle = Rectangle(50, 50, 50, 50, color='red')
circle = Circle((200, 200), 100)

'''game loop'''
run = True

while run:
    time.delay(50)
    window.fill(bg_color)

    # handling events
    for e in event.get():
        if e.type == QUIT:
            run = False

        elif e.type == KEYDOWN: # ngeklik keyboard
            if e.key == K_LEFT:
                player.x_speed = -5
            elif e.key == K_RIGHT:
                player.x_speed = 5
            elif e.key == K_UP:
                player.y_speed = -5
            elif e.key == K_DOWN:
                player.y_speed = 5

        elif e.type == KEYUP: # lepasin tombol
            if e.key == K_UP:
                player.y_speed = 0
            elif e.key == K_DOWN:
                player.y_speed = 0
            elif e.key == K_LEFT:
                player.x_speed = 0
            elif e.key == K_RIGHT:
                player.x_speed = 0

    # objects
    wall1.reset()
    player.reset()
    rectangle.draw()
    circle.draw()

    player.update()

    display.update()
```

```python
from pygame import *

bg_width = 700
bg_height = 500
window = display.set_mode((bg_width, bg_height))
display.set_caption('Maze Game')

bg_color = (194, 236, 254)

'''class parent'''
class GameSprite(sprite.Sprite):
    def __init__(self, img, width, height, x, y):
        sprite.Sprite.__init__(self)
        self.image = transform.scale(image.load(img), (width, height))
        self.rect = self.image.get_rect()
        # position
        self.rect.x = x # x coordinate
        self.rect.y = y # y coordinate

    # placing object on the screen
    def reset(self):
        window.blit(self.image, (self.rect.x, self.rect.y))

# class player
class Player(GameSprite):
    def __init__(self, img, width, height, x, y, x_speed, y_speed):
        GameSprite.__init__(self, img, width, height, x, y)
        self.x_speed = x_speed
        self.y_speed = y_speed

    # movement
    def update(self):
        # horizontal movement
        self.rect.x += self.x_speed
        platform_touched = sprite.spritecollide(self, walls, False) # sprite and group
        if self.x_speed > 0: # gerak ke kanan
            for p in platform_touched:
                self.rect.right = min(self.rect.right, p.rect.left)
        elif self.x_speed < 0: # gerak ke kiri
            for p in platform_touched:
                self.rect.left = max(self.rect.left, p.rect.right)

        # vertical movement
        self.rect.y += self.y_speed
        platform_touched = sprite.spritecollide(self, walls, False) # sprite and group
        if self.y_speed > 0: # gerak ke atas
            for p in platform_touched:
                self.rect.bottom = min(self.rect.bottom, p.rect.top)
        elif self.y_speed < 0: # gerak ke bawah
            for p in platform_touched:
                self.rect.top = max(self.rect.top, p.rect.bottom)


# objects
wall1 = GameSprite('platform2.png', 400, 50, 100, 200)
wall2 = GameSprite('platform2.png', 50, 400, 350, 100)
player = Player('hero.png', 80, 80, 50, 400, 0, 0)
final = GameSprite('images/pac-1.png', 80, 80, 600, 400)

# group of sprites
walls = sprite.Group()
walls.add(wall1)
walls.add(wall2)

# win / lose
win = transform.scale(image.load('images/thumb.jpg'), (bg_width, bg_height))
font.init()
font = font.SysFont('Arial', 40)
win_text = font.render('YOU WIN!', True, (0,0,0))

'''game loop'''
run = True
finish = False

while run:
    time.delay(50)

    # handling events
    for e in event.get():
        if e.type == QUIT:
            run = False

        elif e.type == KEYDOWN: # ngeklik keyboard
            if e.key == K_LEFT:
                player.x_speed = -5
            elif e.key == K_RIGHT:
                player.x_speed = 5
            elif e.key == K_UP:
                player.y_speed = -5
            elif e.key == K_DOWN:
                player.y_speed = 5

        elif e.type == KEYUP: # lepasin tombol
            if e.key == K_UP:
                player.y_speed = 0
            elif e.key == K_DOWN:
                player.y_speed = 0
            elif e.key == K_LEFT:
                player.x_speed = 0
            elif e.key == K_RIGHT:
                player.x_speed = 0

    # finish
    if finish == False:
        window.fill(bg_color) # drawing background
        # objects
        # wall1.reset()
        # wall2.reset()
        walls.draw(window)
        player.reset()
        final.reset()

        player.update()

        # collision handling --> win
        if sprite.collide_rect(player, final): # sprite-sprite
            finish = True
            window.blit(win, (0,0))
            window.blit(win_text, (0,0))

    display.update()
```
